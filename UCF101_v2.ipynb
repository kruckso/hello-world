```python
import cv2
import os

src_path = '/content/poc_dataset/'  # Path where your .avi videos are stored
dest_path = '/content/UCF101_frames/'  # Path where extracted frames will be stored
N = 5  # Extract every N frames

if not os.path.exists(dest_path):
    os.makedirs(dest_path)

# Loop through each video file
for class_folder in os.listdir(src_path):
    class_path = os.path.join(src_path, class_folder)
    dest_class_path = os.path.join(dest_path, class_folder)

    if not os.path.exists(dest_class_path):
        os.makedirs(dest_class_path)

    for video_file in os.listdir(class_path):
        video_path = os.path.join(class_path, video_file)
        video_name = video_file.split('.')[0]
        dest_video_path = os.path.join(dest_class_path, video_name)

        if not os.path.exists(dest_video_path):
            os.makedirs(dest_video_path)

        # Read the video
        cap = cv2.VideoCapture(video_path)
        frame_count = 0

        while True:
            ret, frame = cap.read()
            if not ret:
                break

            # Save every Nth frame as an image
            if frame_count % N == 0:
                frame_filename = os.path.join(dest_video_path, f"img_{frame_count:04d}.jpg")
                cv2.imwrite(frame_filename, frame)

            frame_count += 1

        cap.release()

print("Frame extraction completed!")
```
